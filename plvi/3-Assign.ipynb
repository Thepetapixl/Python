Importing the libraries

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
file = "breast_dataset.csv"
data = pd.read_csv(file)
data.head()
```

```python
pd.isnull(data).sum()
```

```python
data = data.drop(["id","Unnamed: 32"], axis=1)
data
```

```python
data["diagnosis"].unique()
```

```python
X = data.iloc[:,1:].values
y = data.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
y = lb.fit_transform(y)
y[:5]
```

```python
from sklearn.model_selection import train_test_split
X_tr, X_te, y_tr, y_te = train_test_split(X,
                                          y,
                                          test_size=0.2,
                                          random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_tr[:,:] = sc.fit_transform(X_tr[:,:])
X_te[:,:] = sc.fit_transform(X_te[:,:])
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_tr, y_tr)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_tr = confusion_matrix(y_tr, classifier.predict(X_tr))
print(cm_tr)
print(accuracy_score(y_tr, classifier.predict(X_tr))*100)
```

```python
sb.heatmap(cm_tr,annot=True)
plt.show()
```

```python
y_pred = classifier.predict(X_te)
comp = pd.DataFrame(np.concatenate((y_te.reshape(len(y_te),1),
                                   y_pred.reshape(len(y_pred),1)),
                                   1),
                    columns=["Actual","Predicted"])
comp
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_te = confusion_matrix(y_te, y_pred)
cm_te
```

```python
sb.heatmap(cm_te,annot=True)
plt.show()
```
