MLP Classifier and Regressor

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "Sonar.csv"
dataset = pd.read_csv(file)
dataset.head()
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print("X=",X[:2])
print("y=",y[:2])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier
classifier = MLPClassifier(hidden_layer_sizes=(100), 
                           max_iter=300,
                           activation = 'relu',
                           solver='sgd',
                           random_state=0,
                           verbose=True)
a = classifier.fit(X_train,y_train)
```

```python
plt.plot(a.loss_curve_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm_train = confusion_matrix(y_train, classifier.predict(X_train))
print(cm_train)
print(accuracy_score(y_train, classifier.predict(X_train))*100)
```

```python
from seaborn import heatmap
heatmap(cm_train, annot=True)
```

```python
y_pred = classifier.predict(X_test)
comp = pd.DataFrame(np.concatenate((y_test.reshape(len(y_test),1),
                                   y_pred.reshape(len(y_pred),1)),
                                   1), 
                                   columns=["actual","predicted"])

comp
```

```python
cm_test = confusion_matrix(y_test,y_pred)
print(cm_test)
print(accuracy_score(y_test,y_pred)*100)
```

```python
#Regressor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file1 = "hourly_wages_data.csv"
dataset1 = pd.read_csv(file1)
dataset1.head()
```

```python
dataset1.info()
```

```python
X = dataset1.drop(["wage_per_hour"],axis=1).values
y = dataset1["wage_per_hour"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:,:] = sc.fit_transform(X_train[:,:])
X_test[:,:] = sc.fit_transform(X_test[:,:])
```

```python
from sklearn.neural_network import MLPRegressor
regressor = MLPRegressor(hidden_layer_sizes=(100),
                         activation='relu',
                         solver='sgd',
                         max_iter=300,random_state=0,verbose=True)
regressor.fit(X_train, y_train)
```

```python
print("Train Accuracy",regressor.score(X_train, y_train)*100,"%")
```

```python
print("Train Accuracy",regressor.score(X_test, y_test)*100,"%")
```
